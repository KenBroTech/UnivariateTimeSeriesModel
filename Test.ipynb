# Times Series Prediction Using RNN

### What is Recurrent Neural Network?
- A recurrent neural network (RNN) is a type of artificial neural network which uses sequential data or time series data. 

- They are distinguished by their <b>“memory”</b> as they take information from prior inputs to influence the current input and output.

### Types of Recurrent Neural Network Architecture

- Bidirectional recurrent neural networks (BRNN)

- Long short-term memory (LSTM)

- Gated recurrent units (GRUs)

## Data pre-processing

```python
import numpy as np

def preprocess_data(series_data, n_features):
    X, y = [], []
    for i in range(len(series_data)):
        input_series_end = i + n_features
        if input_series_end > len(series_data)-1:
            break
        seq_x, seq_y = series_data[i:input_series_end], series_data[input_series_end]
        X.append(seq_x)
        y.append(seq_y)
    return np.array(X), np.array(y)
```

```python
series_data = [100, 200, 320, 400, 530, 370, 600, 710, 890]
n_features = 3
preprocess_data(series_data, n_features)
```

```python
X, y = preprocess_data(series_data, n_features)
```

```python
# Import RNN libraries - Models -> Sequential, Layers -> Dense, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
```

```python
# Build RNN Model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
model = Sequential()
model.add(LSTM(50, activation='relu', input_shape=input_shape, return_sequences=True))
model.add(LSTM(40, activation='relu'))
model.add(Dense(1))
```

```python
# Check RNN Model Summary
model.summary()
```

```python
# Check RNN Model initial_weights
model.get_weights()
```

```python
# Compile model
model.compile(optimizer='adam', loss='mse')
```

```python
# Fit model
model.fit(X, y, epochs=100)
```

```python
# Save model
model.save('savedModel.h5')
```
